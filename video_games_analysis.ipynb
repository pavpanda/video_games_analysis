```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = ["Minecraft", "GTA V", "Tetris (EA)", "Wii Sports", "PUBG", "Super Mario Bros."]
y = [200000000, 130000000, 100000000, 82900000, 70000000, 48240000]
plt.style.use('ggplot')
plt.bar(x, y)
plt.title("Sales by Video game")
plt.ylabel("Sales (100 million)")
plt.xlabel("Video game")
plt.xticks(rotation="45")
plt.savefig("top6games.png", bbox_inches="tight")
```
